# Implementing Clustering Algorithms:

```python
from sklearn.cluster import KMeans
d1=[]
for i in range(1,12):
    kmeans=KMeans(n_clusters=i).fit(scaled_df)
    d1.append(kmeans.inertia_)
plt.plot(range(1,12),d1)
plt.title('The Elbow Method')
plt.xlabel('No.of Clusters')
plt.ylabel('d1')
plt.show()
```

```python

k_clusters=KMeans(n_clusters=3)
k_clusters.fit(scaled_df)
```

```python
kmeans_labels=k_clusters.labels_
kmeans_labels
```

```python
#df=pd.Series(kmeans_labels)

#data['cluster no.']=df
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
#to find epsilon value
nn=NearestNeighbors(n_neighbors=10)
nn.fit(scaled_df)
distances,idices=nn.kneighbors(scaled_df)
distances=np.sort(distances,axis=0)
distances=distances[:,1]
```

```python
plt.plot(distances)
plt.title('K-distance Graph')
plt.xlabel('data points')
plt.ylabel('Epsilon')
plt.show()
```

```python
dbscan=DBSCAN(eps=2,min_samples=9)
dbscan_labels=dbscan.fit_predict(scaled_df)
dbscan_labels
```

# Visualization:

```python
#Visualization for k means clustering
plt.scatter(scaled_df[:,0],scaled_df[:,1],c=kmeans_labels,cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature-1')
plt.ylabel('Feature_2')
plt.show()
```

```python
plt.scatter(scaled_df[:,0],scaled_df[:,1],c=dbscan_labels)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature-1')
plt.ylabel('Feature_2')
plt.show()
```
