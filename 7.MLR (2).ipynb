```python
import numpy as np
import pandas as pd
df = pd.read_csv("ToyotaCorolla - MLR.csv")
df
```

```python
Y=df["KM"]
X=df[["HP"]]
```

```python
#X=df[["HP","cc"]]
#X=df[["HP","cc","Cylinders"]]
```

```python
# fit the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
y_pred = model.predict(X)
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(Y,y_pred)
print("Mean squared error:", np.round(mse,2)) 
print("root Mean squared error:", np.round (np.sqrt(mse),2))
score = r2_score(Y,y_pred)
print("R square:",np.round(score,2))
```

2.What techniques can be used to address multicollinearity in multiple linear regression?
A. Techniques to handle multicollinearity (High correlation between predictors)
    1.remove one of the correlated variables
    2.combine correlated variables
    3.use feature selection methods
      forward selection /backward elimination / stepwise regression 
    4.Regularization 
    *Ridge (L1)
    *Lasso (L2)
    *elastic net(L1&L2)
    5.Principal component Regression(PCR)
    6.Standardizing predictors
    7.Collect more data
