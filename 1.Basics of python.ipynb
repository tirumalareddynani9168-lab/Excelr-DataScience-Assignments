# PRIME NUMBERS

```python
#Write a Python program that checks whether a given number is prime or not.A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself?
```

```python
number = int(input("Enter any number: "))
if number >1:
    for i in range(2,number):
        if (number % i) ==0:
          print(number,"is not a prime number")
          break
    else:
        print(number,"is a prime number")
```

```python
# Exercise 4:
# Problem Statement (Question): Hospital Billing System
#A hospital charges its patients based on the following:
#•	Room charges per day: ₹2,000
#•	Doctor consultation fee (flat): ₹1,500
#•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
#•	Medicine charges: Total cost of medicines provided
#•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
#Write a Python program to calculate the final bill for a patient who:
#•	Stayed for 4 days
#•	Had 3 lab tests
#•	Had medicine charges of ₹2,400 
```

```python
days_stayed = 2
doctor_fee = 1500
room_charge_per_day = 2000
lab_charge_per_test = 300

num_lab_tests = int(input("Enter the number of lab tests: "))
medicine_cost = float(input("Enter the total medicine cost: "))

room_charges = days_stayed * room_charge_per_day
lab_charges = num_lab_tests * lab_charge_per_test 

total_bill = room_charges + doctor_fee + lab_charges + medicine_cost
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount
print("\n--- Hospital Bill ---")
print(f"Room Charges ({room_charge_per_day} x {days_stayed} days): {room_charges}")
print(f"Doctor Consultation Fee: {doctor_fee}")
print(f"Lab Test Charges ({lab_charge_per_test} x {num_lab_tests} tests): {lab_charges}")
print(f"Medicine Charges: {medicine_cost}")
print(f"Discount: {discount}")
print(f"Final Bill Amount: {final_bill}")
```
