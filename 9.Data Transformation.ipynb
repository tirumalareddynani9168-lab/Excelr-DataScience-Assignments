```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns
df_encoded = df.copy()

for col in categorical_cols:
    if df[col].nunique() < 5:
      
        dummies = pd.get_dummies(df_encoded[col], prefix=col, drop_first=True)
        df_encoded = pd.concat([df_encoded, dummies], axis=1)
       
        df_encoded = df_encoded.drop(columns=[col])

print("DataFrame after One-Hot Encoding for selected categorical variables:")
print(df_encoded.head())
```

```python
from sklearn.preprocessing import LabelEncoder
remaining_categorical_cols = df_encoded.select_dtypes(include='object').columns

df_labeled = df_encoded.copy()

for col in remaining_categorical_cols:
    le = LabelEncoder()
    df_labeled[col] = le.fit_transform(df_labeled[col])
    

print("\nDataFrame after Label Encoding (first 5 rows):")
print(df_labeled.head())
```

```python
# Create a copy to add new features to the labeled DataFrame
df_features = df_labeled.copy()

# Feature 1: Net_Capital_Movement
df_features['Net_Capital'] = df_features['capital_gain'] 
print("Added 'Net_Capital' feature.")

# Feature 2: High_Education_Hours_Interaction
df_features['High_Education'] = df_features['education_num'] 
print("Added 'High_Education' feature.")

print("\nDataFrame after adding new features:")
print(df_features.head())
```

```python
import numpy as np

initial_skewness = df_features['capital_gain'].skew()
print(f"Initial skewness of 'capital_gain': {initial_skewness:.2f}")

# Apply log1p transformation
df_features['capital_gain_log'] = np.log1p(df_features['capital_gain'])

# Check skewness after transformation
transformed_skewness = df_features['capital_gain_log'].skew()
print(f"Skewness of 'capital_gain_log' after transformation: {transformed_skewness:.2f}")

print("\nDataFrame after adding 'capital_gain_log' (first 5 rows):")
print(df_features[['capital_gain', 'capital_gain_log']].head())
```
