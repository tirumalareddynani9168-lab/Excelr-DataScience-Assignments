```python
import numpy as np
from scipy import stats
df = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
df
```

```python
import numpy as np
from scipy import stats

# Calculate sample mean and standard deviation
sample_mean = np.mean(df)
sample_std = np.std(df, ddof=1) 
n = len(df)

# Confidence level and alpha
confidence_level = 0.99
alpha = 0.5 - confidence_level

# Degrees of freedom
df_dof = n - 1

# T-critical value for a two-tailed test
t_critical = stats.t.ppf(1 - alpha/2, df_dof)

# Margin of Error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence Interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Degrees of Freedom: {df_dof}")
print(f"T-critical value ({confidence_level*100}%): {t_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"{confidence_level*100}% Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")
```
