```python
import numpy as np
df_numerical = df.select_dtypes(include=np.number)
print(df_numerical.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numerical_cols = df_numerical.select_dtypes(include=['float64', 'int64']).columns

for column in numerical_cols:       # Create histograms
    plt.figure(figsize=(8, 6))
    sns.histplot(df_numerical[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_for_bar_charts = ['DS', 'Tendency', 'NSP']

if not categorical_for_bar_charts:
    print("No specific categorical columns were identified for bar charts.")
else:
    print(f"Generating bar charts for categorical columns: {categorical_for_bar_charts}")
    for column in categorical_for_bar_charts:
        plt.figure(figsize=(8, 5))
        sns.countplot(data=df, x=column, palette='viridis')
        plt.title(f'Bar Chart of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='LB', y='ASTV')
plt.title('Scatter Plot of LB vs ASTV')
plt.xlabel('LB (Baseline FHR)')
plt.ylabel('ASTV (Abnormal Short Term Variability)')
plt.grid(True)
plt.show()
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12, 10))  # Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
