# Data Preprocessing & Feature extraction

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["genre"].mode()
```

```python
df["genre"]=df["genre"].fillna(df["genre"].mode()[0])

df["rating"]=df["rating"].fillna(0)
```

```python

df.isnull().sum()
```

```python
df1=pd.pivot_table(df,index='anime_id',columns='genre',values='rating')
```

```python
df1.fillna(0,inplace=True)
df1
```

# Recommendation System

```python
from sklearn.metrics.pairwise import cosine_similarity
df_sim=cosine_similarity(df1)
df_sim
```

```python
sim_df=pd.DataFrame(df_sim)
sim_df
```

```python
sim_df.index=df.anime_id.unique()
sim_df.columns=df.anime_id.unique()
sim_df
```

```python
np.fill_diagonal(df_sim,0)
sim_df.idxmax()[0:10]
```
