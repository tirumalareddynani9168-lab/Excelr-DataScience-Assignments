### Boxplots for Outliers Detection

```python
plt.figure(figsize=(16,8))

sns.boxplot(data=df.select_dtypes(include=np.number))

plt.xticks(rotation=90)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

The boxplots reveal the presence of significant outliers in several variables, especially in features such as GDP, Tourism Inbound, Tourism Outbound. These extreme values indicate that some countries have much higher economic or tourism activity compared to others. 

```python
# Select only numerical columns
d1 = df.select_dtypes(include=np.number).columns
```

```python
outlier_count = {}

for col in d1:
    
    Q1 = df[col].quantile(0.25)   
    Q3 = df[col].quantile(0.75)  
    IQR = Q3 - Q1                
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    outlier_count[col] = len(outliers)

outlier_df = pd.DataFrame(outlier_count.items(), columns=['Column','Outlier_Count'])

outlier_df
```

Although several variables show extreme values, these outliers represent real economic and developmental differences among countries rather than data errors. Therefore, instead of removing them, feature scaling techniques are applied to ensure that clustering algorithms handle these variations appropriately while preserving meaningful country-level information.

## Model-1 K-Means Clustering


**Basic Idea of K-Means**

K-Means is an unsupervised learning algorithm used to group similar data points into clusters.

**Working principle:**

* Choose number of clusters (K)
* Randomly initialize K centroids
* Assign each data point to the nearest centroid
* Recalculate centroids (mean of cluster points)
* Repeat until centroids stop changing (convergence)
