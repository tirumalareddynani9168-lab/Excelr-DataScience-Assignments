# Random Forest Model Implementation

```python
features=df.drop('Type',axis=1)
features
```

```python
target=df['Type']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=42)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50,max_features=3)
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)

print(classification_report(y_test,ypred))
```

# Bagging and Boosting Methods:

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
model2=BaggingClassifier(dt,n_estimators=40)
model2.fit(x_train,y_train)
```

```python
from sklearn.ensemble import AdaBoostClassifier
base_estimator=DecisionTreeClassifier()
model3=AdaBoostClassifier(base_estimator,n_estimators=40)
model3.fit(x_train,y_train)
```

```python
y_pred=model3.predict(x_test)

print(classification_report(y_test,y_pred))
```

# additional note:

 1.Explain Bagging and Boosting methods. How is it different from each other.
 A.a)Bagging: Train many models independently (parallel) on different random samples, then average/vote the results. â†’ Reduces overfitting (variance).
   b)Boosting: Train models one after another (sequential), each new model fixes previous mistakes, then combine them. â†’ Improves accuracy (reducebias).
   Main difference:
Bagging = parallel + reduces variance
Boosting = sequential + reduces bias

2.Explain how to handle imbalance in the data.
Resampling: Oversample minority (e.g., SMOTE) or undersample majority.
Use class weights: Set class_weight='balanced' (or cost-sensitive learning).
Try better models: Tree ensembles (Random Forest, XGBoost) often work well with weights.
Use proper metrics: F1-score, Recall, Precision-Recall AUC, not just accuracy.
Tune threshold: Change probability cutoff (not always 0.5) to improve recall/precision.
