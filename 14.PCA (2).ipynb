# Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import scale

norm=scale(df)

norm
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_val=pca.fit_transform(norm)
pca_val
```

```python
var=pca.explained_variance_ratio_

var1=np.cumsum(np.round(var,4)*100)
var
```

# Clustering with Original Data

```python
from sklearn.cluster import KMeans
d1=[]
for i in range (1,15):
    kmeans=KMeans(n_clusters=i).fit(norm)
    d1.append(kmeans.inertia_)
```

```python
plt.plot(range(1,15),d1)
plt.title('The Elbow Method')
plt.xlabel('No.of Clusters')
plt.ylabel('d1')
plt.show()
```

```python
k_clusters=KMeans(n_clusters=3)
k_clusters.fit(norm)
```

```python
labels=k_clusters.predict(norm)
labels
```

```python
plt.scatter(norm[:,12],norm[:,8],c=labels,cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature-1')
plt.ylabel('Feature_2')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
kmeans_silhouette_score = silhouette_score(pca_val, labels)
print("K-Means Silhouette Score:", kmeans_silhouette_score)
```

# Clustering with PCA Data

```python
comparing_cluster=pd.DataFrame({'ORIGINAL DATA':kmeans_silhouette_score,'PCA  DATA':kmeans_silhouette_score},index=(0,))
comparing_cluster
```
